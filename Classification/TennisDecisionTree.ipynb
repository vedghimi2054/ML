```python
import pandas as pd
df=pd.read_csv('tennis.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_outlook=LabelEncoder()
le_temp=LabelEncoder()
le_humidity=LabelEncoder()
le_windy=LabelEncoder()
le_play=LabelEncoder()
```

```python
df['outlook_n']=le_outlook.fit_transform(df['outlook'])
df['temp_n']=le_temp.fit_transform(df['temp'])
df['humidity_n']=le_humidity.fit_transform(df['humidity'])
df['windy_n']=le_windy.fit_transform(df['windy'])
df['play_n']=le_play.fit_transform(df['play'])
```

```python
df.head()
```

```python
df_n=df.drop(['outlook','temp','humidity','windy','play'],axis='columns')
```

```python
df_n
```

```python
inputs=df_n.drop('play_n',axis='columns')
target=df['play_n']
```

```python
inputs
```

```python
target
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier(criterion='entropy')#entroy used for information gain
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

```python
model.predict([[0,1,0,0]])
```
