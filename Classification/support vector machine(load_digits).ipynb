```python
df=pd.DataFrame(digits.data,digits.target)
```

```python
df.head()
```

```python
df['target_names']=digits.target
df.head(20)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('target_names',axis='columns'), df.target_names, test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf',C=100,gamma='auto')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
